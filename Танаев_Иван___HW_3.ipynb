```python
class Driver():
    def __init__(self, first_name: str, last_name: str, driver_id: int, radio_number: str, wage: float):
        self.first_name = first_name
        self.last_name = last_name
        self.driver_id = driver_id
        self.radio_number = radio_number
        self.wage = wage

    def change_wage(self, change):
        if self.wage != 0:
            self.wage += change
        else:
            return 'The driver is fired'

    def fire(self):
        self.first_name = 'fired'
        self.last_name = 'fired'
        self.driver_id = -1
        self.radio_number = 'fired'
        self.wage = 0


class car():
    def __init__(self, x: float, y: float, firm: str, model: str, capacity: float, age: float, driver: Driver = None):
        self.x = x
        self.y = y
        self.driver = driver
        self.firm = firm
        self.model = model
        self.capacity = capacity
        self.age = age

    def assign_driver(self, driver: Driver):
      if driver.first_name == 'fired':
          return 'You cannot assign a fired driver'
      else:
        self.driver = driver

    def remove_driver(self):
        self.driver = None

    def get_capacity_category(self):
        if self.capacity < 90:
            return "Малой грузоподъемности"
        elif self.capacity < 180:
            return "Средней грузоподъемности"
        else:
            return "Большой грузоподъемности"

    def help(self):
      if self.age >= 20:
        return 'Help of service is required'
      else:
        return 'No service needed'


class standartcar(car):
    def __init__(self, x: float, y: float, firm: str, model: str, capacity: float, cargo: str, age: float, driver: Driver = None):
        super().__init__(x, y, firm, model, capacity,age, driver)
        self.cargo = cargo

    def change_cargo(self, new_cargo: str):
        self.cargo = new_cargo


class servicecar(car):
    def __init__(self, x: float, y: float, firm: str, model: str, age:float, driver: Driver = None):
        super().__init__(x, y, firm, model, 0, age, driver)


class operator():
    def __init__(self, name: str):
        self.name = name

    def assign_driver_to_car(self, driver: Driver, machine: car):
        result = machine.assign_driver(driver)
        if result:
            print(result)

    def remove_driver_from_car(self, machine: car):
        machine.remove_driver()
```
